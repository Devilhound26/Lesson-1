### Variables
- Variables are used to store information to be referenced and manipulated in a computer program
- They also provide a way of labeling data with a descriptive name, so our programs can be understood more clearly by the reader and ourselves
- It is helpful to think of variables as containers that hold information (File cabinet analogy)

```python
# Variables with Strings
# A function that printer a string to the console
print("Hello World")
# Let change the print statement to my name
print("Greg Walsh")
# What if I want to change my name?
print("Gregory Walsh")
# Make name a String Variable
# Variable_name = Variable_value
# Filing Cabinet Analogy
# Folder Name = Folder Contents
name = "Gregory Walsh" # String Variable
# String name = "Gregory Walsh" // Java
# in Python we don't need to specify the type of the variable when we create it
# this is helpful for when you want to know the type of a variable
print(type(name))
# now that we have a variable for our name, we can use it in our print statement
print("The variable name is:", name) # String Concatenation
# String Concatenation - combining strings together
# How do I change the variable name?
name = "Greg" # change the value of the variable name to my first name
print("The variable name is:", name)
# Copies of variables
# When you copy a variable, you are making a copy of the value
first_name = name
# This look like it is copying the variable name into first_name
# first_name = "Greg"
print("The variable first_name is:", first_name)

last_name = "Walsh" # String Variable


# How would I print my full name?
print("Greg Walsh")
print("Gregory Walsh")

# String Concatenation
# + 
print("Greg" + "Walsh")
# Notice there is not space 
print("Greg" + " " + "Walsh")

# What if I want to print my name using the variables we just made?
print(first_name + last_name)
# notice we are still missing a space. This is because we did not add a space to the end of the first_name variable
print(first_name + " " + last_name)
```

```python
# Variable with Ints (Integers)

current_age = 26 # Int Variable
phone_number = 1234567890 # Int Variable
# Note: Notice how the variable name has a '_' in it. This is called snake case. It is a common naming convention in Python
# In Java we would use camel case
# phoneNumber = 1234567890 // Java

number_one = 5
number_two = 10


# What if I want to print my age?
print(current_age)
print(type(current_age))

print("Print number_one Plus number_two:")
print(number_one + number_two)

# The + in this isntance is not concatenation, it is addition
# This is because we are adding two ints together
# When you add two ints together, you get an int
# When you add two strings together, you get a string
# When you add a string and an int together, you get an error
# IE: We will try to add a string (name) with an int (number_one)
# Greg + 5

# print(name + number_one)

# this will create a TypeError
# TypeError: can only concatenate str (not "int") to str

```

```python
## Python Arithmetic Operators
#  + - * / % ** // 
#  + addition
#  - subtraction
#  * multiplication
#  / division
#  % modulus (remainder)
#  ** exponent
#  // floor division (rounds down)
```

```python
# What if I want to print my age in 10 years?

future_age = current_age + 10

# variable age = 26
# Remember, the = sign is the assignment operator
# It assigns the value on the right to the variable on the left
# So, this statement is saying:
# Assign the value of age + 10 to the variable age
# remember variables can change

# 26 = 26 + 10

print("My age in 10 years is: ", future_age)

# What if I ant to print my age from 10 years ago? 
# First I need to assign the value of age back to the origianl value

past_age = current_age - 10 
print("My age 10 years ago is: ", past_age)

print(future_age)
print(past_age)
print(current_age)

# What is my age times 2?
print(current_age * 2)
# notice I am not making a new variable here like I was for future_age and past_age
# I am just printing the value of current_age * 2
# There will be time you do not need to create variables for everything

# What is my age divided by 2?
print(current_age / 2)
# notice the output is a float
# this is because we are dividing an int by an int

# Some other division examples:
# when you divide an int by an int, you get a float
# when you divide a float by an int, you get a float
# when you divide an int by a float, you get a float
# when you divide a float by a float, you get a float


# What is my age modulus 2?
# What is Modulus?
# Modulus is the remainder of a division problem
# 10 / 2 = 5
# 10 % 2 = 0 "10 Mod 2 is 0"
# 10 / 3 = 3.3333333333333335
# 10 % 3 = 1 "10 Mod 3 is 1"

print(current_age % 5)
# 5 goes into 26 5 times with a remainder of 1 
# 5 10 15 20 25 (26)
# 5 with 1 left over
# 1 is the remainder

exponent_test = 2

# 2 to the 2nd exponent
print(exponent_test ** 2)


```
